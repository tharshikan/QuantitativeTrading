# Simulating two stocks


```python
np.random.seed(100)
# use returns to create a price series


returns = pd.Series(np.random.normal(0, 1, 1000))

noise1 = pd.Series(np.random.normal(0, 0.3, 1000))
noise2 = pd.Series(np.random.normal(0, 0.4, 1000))
drift1 = 50
drift2 = 60

r_cum_1 = pd.Series(returns.cumsum(), name='rcum')+noise1+drift1
r_cum_2= pd.Series(returns.cumsum(), name='rcum')+ noise2+drift2
noise_cum = pd.Series(noise1.cumsum(), name='rcum')

r_cum_with_noise_1 = r_cum_1+noise1

# rcum2= pd.Series(returns.cumsum(), name='rcum with drift') + drift
pd.concat([r_cum_1, r_cum_2], axis=1).plot(figsize=(15, 6))
```
